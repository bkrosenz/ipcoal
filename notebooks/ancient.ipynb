## IPCOAL WITH ANCIENT SAMPLES

```python
import ipcoal
import toytree
import msprime as ms
```

## Species tree

```python
tre = toytree.rtree.rtree(5, treeheight=1e6, seed=33333)
tre.draw(ts='p', edge_type='p');
```

```python
self = ipcoal.Model(tre, Ne=1e4, nsamples=2)
self.sim_loci(1, 15000)
self.infer_gene_trees()
self.df.head()
```

## Genealogy

```python
gtre = toytree.tree(self.df.genealogy[0]).root(regex="r0")
gtre.draw(ts='c', tip_labels=True, edge_type='p');
```

## Inferred

```python
itre = toytree.tree(self.df.inferred_tree[0]).root(regex="r0")
itre.draw(ts='c', tip_labels=True, edge_type='p');
```

### PROBLEMS WITH ADMIXTURE

```python
tre = toytree.rtree.rtree(5, treeheight=1e6, seed=33333)
tre.draw(ts='p', edge_type='p');
```

```python
self = ipcoal.Model(tre, Ne=1e4, admixture_edges=[(0, 2, 0.5, 0.5)])
self.sim_trees(1, 15000)
```
